```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import csv
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
```

```python
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path="./chromedriver")
```

```python
title = []
genre = []
publisher = []
author = []
isbn = []
onelineSummary = []
summary = []
image = []
detailImage = []
isPublish = []
```

```python
import time
```

```python
for i in range(100, 200):
    url = "http://www.joara.com/literature/view/book_list.html?page_no={}&bookpart=&sl_type=&sl_chkcost=&sl_category=etc&sl_search=&sl_keyword=&sl_chk=&sl_minchapter=&sl_maxchapter=&sl_redate=&sl_orderby=&sl_othercategory=&list_type=normal&sub_category=".format(i)
    driver.get(url)
    time.sleep(2)
    processJoara()
    time.sleep(2)
```

```python
def processJoara():
    #image
    elements = driver.find_elements_by_css_selector('[alt="표지 이미지"]')
    for el in elements:
        image.append(el.get_attribute("src"))

    #title
    elements = driver.find_elements_by_xpath("//*[@id='content']/section/table/tbody/tr/td/div/p/a")
    for el in elements:
        text = el.get_attribute("text")
        if (text != '' and text != "\n\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t"):
            title.append(text)

    #author
    elements = driver.find_elements_by_xpath("//*[@id='content']/section/table/tbody/tr/td/div[1]/p[2]/span[1]/span")
    for el in elements:
        text = el.get_attribute("innerHTML")
        if (text != ''):
            author.append(text.replace(" ", "").replace("\n",""))

    #summary
    elements = driver.find_elements_by_xpath("//*[@id='content']/section/table/tbody/tr/td/div[2]/p[1]")
    for el in elements:
        text = el.get_attribute("innerHTML")
        if (text != ''):
            text = (text.replace("\n","").replace("\t","").replace('\\','').replace("&lt;","").replace("&gt;",""))
            summary.append(text)
```

```python
genre = [6] * 1935
onelineSummary = []
```

```python
for i in range(0, len(summary)):
    if (len(summary[i]) >= 25):
        onelineSummary.append(summary[i][:25])
    else:
        onelineSummary.append(summary[i])
```

```python
detailImage = image
isPublish = ['Y'] * 1935
```

```python
publisher = [""] * 1935
isbn = [""] * 1935
```

```python
print(len(title), len(genre), len(publisher), len(author), len(isbn), len(onelineSummary), len(summary), len(image), len(detailImage), len(isPublish))
```

```python
#export csv
rows = zip(title,genre,publisher,author,isbn,onelineSummary,summary,image,detailImage,isPublish)

with open("joara.csv", "w", encoding='utf-8') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
