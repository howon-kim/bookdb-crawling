```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import csv
import time
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path="./chromedriver.exe")
```

```python
title = []
genre = []
publisher = []
author = []
isbn = []
onelineSummary = []
summary = []
image = []
detailImage = []
isPublish = []
```

```python
url = "https://page.kakao.com/ranking/today?categoryUid=11&subCategoryUid=117"
driver.get(url)
```

```python
#title
elements = driver.find_elements_by_xpath("//*[@id='root']/div/div/div/div/a/li/div/div/div[1]/div/div")
for el in elements:
    print(el.text)
```

```python
#author
elements = driver.find_elements_by_xpath("//*[@id='root']/div/div/div/div/a/li/div[3]/div/div[2]/div/div[3]")
for el in elements:
    print(el.text)
```

```python
#image
elements = driver.find_elements_by_xpath("//*[@id='root']/div/div/div/div/a/li/div[2]/div[1]/img")
for el in elements:
    print(el.get_attribute("src"))
```

```python
def processJoara():
    #image
    elements = driver.find_elements_by_css_selector('[alt="표지 이미지"]')
    for el in elements:
        image.append(el.get_attribute("src"))

    #title
    elements = driver.find_elements_by_xpath("//*[@id='content']/section/table/tbody/tr/td/div/p/a")
    for el in elements:
        text = el.get_attribute("text")
        if (text != '' and text != "\n\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t"):
            title.append(text)

    #author
    elements = driver.find_elements_by_xpath("//*[@id='content']/section/table/tbody/tr/td/div[1]/p[2]/span[1]/span")
    for el in elements:
        text = el.get_attribute("innerHTML")
        if (text != ''):
            author.append(text.replace(" ", "").replace("\n",""))

    #summary
    elements = driver.find_elements_by_xpath("//*[@id='content']/section/table/tbody/tr/td/div[2]/p[1]")
    for el in elements:
        text = el.get_attribute("innerHTML")
        if (text != ''):
            text = (text.replace("\n","").replace("\t","").replace('\\','').replace("&lt;","").replace("&gt;",""))
            summary.append(text)
```

```python
genre = [6] * 1935
onelineSummary = []
```

```python
for i in range(0, len(summary)):
    if (len(summary[i]) >= 25):
        onelineSummary.append(summary[i][:25])
    else:
        onelineSummary.append(summary[i])
```

```python
detailImage = image
isPublish = ['Y'] * 1935
```

```python
publisher = [""] * 1935
isbn = [""] * 1935
```

```python
print(len(title), len(genre), len(publisher), len(author), len(isbn), len(onelineSummary), len(summary), len(image), len(detailImage), len(isPublish))
```

```python
#export csv
rows = zip(title,genre,publisher,author,isbn,onelineSummary,summary,image,detailImage,isPublish)

with open("joara.csv", "w", encoding='utf-8') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
