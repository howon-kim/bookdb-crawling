```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import csv
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
```

```python
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path="./chromedriver.exe")
```

```python
title = []
genre = []
publisher = []
author = []
isbn = []
onelineSummary = []
summary = []
image = []
detailImage = []
isPublish = []
```

```python
url = "https://novel.naver.com/webnovel/ranking.nhn?genre=103&periodType=DAILY"
driver.get(url)
```

```python
# div.ranking_wrap_left 먼저 다음에 div.ranking_wrap_right. left 가져올때 이미지 다 가져옴. right에서 이미지 할 필요 없음
elements = driver.find_elements_by_css_selector("#content > div > div.ranking_lst > div.ranking_wrap_right > ul")
```

```python
imageElement = elements[0].find_elements_by_xpath(".//div[@class='pic']")
imageElement2 = imageElement[0].find_elements_by_xpath("//img")
for el in imageElement2:
    image.append(el.get_attribute("src"))
```

```python
authorElement = elements[0].find_elements_by_xpath(".//span[@class='author']")
for el in authorElement:
    author.append(el.text)
```

```python
titleElement = elements[0].find_elements_by_xpath(".//span[@class='tit']")
for el in titleElement:
    title.append(el.text)
```

```python
print(len(title), len(image), len(author))
```

```python
image= image[3:]
```

```python
#export csv
rows = zip(title,author,image)

with open("naver.csv", "w", encoding='euc_kr') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
