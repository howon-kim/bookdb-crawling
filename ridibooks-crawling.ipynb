```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import csv
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
```

```python
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path="./chromedriver")
```

```python
title = []
genre = []
publisher = []
author = []
isbn = []
onelineSummary = []
summary = []
image = []
detailImage = []
isPublish = []
```

```python
url = "https://ridibooks.com/bestsellers/romance_serial?page=8"
driver.get(url)
```

```python
#Book Image
elements = driver.find_elements_by_css_selector("#page_best > div.book_macro_wrapper.js_book_macro_wrapper > div.book_macro_110.book_macro_landscape > div.book_thumbnail_wrapper > div > div > img")
for el in elements:
    if(el.get_attribute("data-original-cover")[:4] == "http"):
        image.append(el.get_attribute("data-original-cover"))
    else:
        image.append("https:" + el.get_attribute("data-original-cover"))
```

```python
#Title
elements = driver.find_elements_by_css_selector("#page_best > div.book_macro_wrapper.js_book_macro_wrapper > div.book_macro_110.book_macro_landscape > div.book_metadata_wrapper > h3 > a > span")
title = [el.text for el in elements]
```

```python
#author
elements = driver.find_elements_by_css_selector("#page_best > div.book_macro_wrapper.js_book_macro_wrapper > div.book_macro_110.book_macro_landscape > div.book_metadata_wrapper > p.book_metadata.author > a")
author = [el.text for el in elements]
```

```python
#publisher
elements = driver.find_elements_by_css_selector("#page_best > div.book_macro_wrapper.js_book_macro_wrapper > div.book_macro_110.book_macro_landscape > div.book_metadata_wrapper > p.book_metadata.publisher > a")
publisher = [el.text for el in elements]
```

```python
#oneline summary
elements = driver.find_elements_by_css_selector("#page_best > div.book_macro_wrapper.js_book_macro_wrapper > div.book_macro_110.book_macro_landscape > div.book_metadata_wrapper > p.book_metadata.meta_description > a")
onelineSummary = [el.text for el in elements]
```

```python
#export csv
rows = zip(image,title,author,publisher,onelineSummary)

with open("ridibooks.csv", "a") as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
